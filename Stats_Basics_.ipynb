1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.

Data is broadly categorized into Qualitative and Quantitative data:

1. Qualitative Data (Categorical Data)
Describes qualities or characteristics that can be observed but not measured numerically.

It groups data into categories.

Examples:

Colors: red, blue, green

Types of animals: dog, cat, bird

Gender: male, female, other

Brands: Nike, Adidas, Puma

Qualitative data can be further divided into:

Nominal: Categories without any order

Ordinal: Categories with a meaningful order

2. Quantitative Data (Numerical Data)
Represents numeric values that measure quantities or amounts.

You can perform arithmetic operations on these values.

Examples:

Height in centimeters: 150, 170, 180

Temperature in °C: 20, 25, 30

Number of children: 1, 2, 3

Test scores: 80, 90, 100

Quantitative data can be further divided into:

Interval: Numeric scales with equal intervals but no true zero

Ratio: Numeric scales with equal intervals and a true zero

2) What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.

Measures of central tendency are statistics that summarize a dataset by identifying its center point. The three main measures are mean, median, and mode. The mean is the average of all values and is best used with continuous data that is normally distributed, like average test scores. The median is the middle value when data is ordered, useful for skewed data or when there are outliers, such as household income. The mode is the most frequent value, ideal for categorical data, like the most popular shoe size in a store. Choosing the right measure depends on the data type and distribution.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?


Concept of Dispersion, Variance, and Standard Deviation:
Dispersion refers to how spread out or scattered the data points are in a dataset. It shows the variability or consistency of the data. Variance measures the average squared difference between each data point and the mean, giving an idea of overall spread. However, since variance is in squared units, standard deviation (the square root of variance) is more commonly used because it is in the same units as the original data. Both help understand how much the data varies around the mean.

4. What is a box plot, and what can it tell you about the distribution of data?


A box plot (or box-and-whisker plot) is a graphical tool that summarizes data using five statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It shows the distribution’s center, spread, and skewness. The “box” covers the interquartile range (middle 50% of data), and the “whiskers” extend to the minimum and maximum values. It also highlights outliers, making it easy to visualize data symmetry and variability.

5. Discuss the role of random sampling in making inferences about populations.

Random sampling is the process of selecting a subset of individuals from a population where each member has an equal chance of being chosen. This method reduces bias and ensures the sample represents the population fairly. It is essential for making valid inferences because it allows statistical techniques to generalize findings from the sample to the entire population, improving the accuracy and reliability of conclusions.

6 Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Skewness measures the asymmetry of a data distribution. If data is symmetric, skewness is zero. Positive skewness (right skew) means the tail is longer on the right side, indicating many low values and few very high values. Negative skewness (left skew) has a longer left tail, meaning many high values and few very low values. Skewness affects interpretation by showing if the mean is pulled away from the median, influencing which measure of central tendency is best used.

 9) Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule)

Normal Distribution Properties:
Symmetrical bell-shaped curve.

Mean = Median = Mode (center).

Defined by mean (μ) and standard deviation (σ).

Tails extend infinitely but never touch the axis.

Total area under curve = 1 (100% probability).

Empirical Rule (68-95-99.7):
68% of data within 1σ of mean.

95% of data within 2σ of mean.

99.7% of data within 3σ of mean.

A random variable is a numerical value that represents the outcome of a random experiment. It assigns a number to each possible outcome in the sample space of a random process.

Types of Random Variables:
Type	Description	Example

Continuous Random Variable	Takes any value within a continuous range or interval.	Height of a person (e.g., 160.5 cm, 172.3 cm)
